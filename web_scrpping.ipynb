```python
import pandas as pd
# Initialize lists

cards=soup.find_all('div',class_='styled__StyledRestaurantGridCard-sc-fcg6mi-0 lgOeYp')
def extract_cuisines_and_addresses(cu):
    cuisines = []
    addresses = []
    
    for index, item in enumerate(cu):
        text = item.get_text(strip=True)  # Extract and clean the text
        if index % 2 == 0:  # Even index, cuisine
            cuisines.append(text)
        else:  # Odd index, address
            addresses.append(text)
    
    return cuisines, addresses




name = []
ratings = []
cuisines = []
pricing = []
address = []
distance = []
extra_info = []
closing = []

# Extract the data
for card in cards:
    # Get the name of the restaurant
    name_elements = card.find_all('div', class_="sc-beySbM iKLEMo")
    name.append(', '.join([elem.get_text(strip=True) for elem in name_elements]))

    # Get the rating
    rating_elements = card.find_all('span', class_="sc-beySbM jAtOQO")
    ratings.append(', '.join([elem.get_text(strip=True) for elem in rating_elements]))
    
    
    # Get the pricing
    pricing_elements = card.find_all('div',class_='sc-beySbM fsIdyU sc-qZqnM dSTtWe')
    pricing.append(', '.join([elem.get_text(strip=True) for elem in pricing_elements])or None)
    
    
   # Get the cuisine & addresss
    cu = card.find_all('div', class_="sc-beySbM bRTXBF")
    extracted_cuisines, extracted_addresses = extract_cuisines_and_addresses(cu)
    
    cuisines.append(', '.join(extracted_cuisines))
    address.append(', '.join(extracted_addresses))
    
    # Get the distance
    distance_elements = card.find_all('div', class_="sc-beySbM brneVe")
    distance.append(', '.join([elem.get_text(strip=True) for elem in distance_elements]))

# Create a DataFrame
dist = pd.DataFrame({
    'name_r': name,
    'ratings': ratings,
    'cuisines': cuisines,
    'pricing': pricing,
    'address': address,
    'distance': distance,
   
})

# Display the DataFrame
swiggy_surat=pd.DataFrame(dist)
 
```
